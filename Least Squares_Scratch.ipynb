```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from numpy.linalg import inv

xm,ym = np.loadtxt("C:\Semester 2\ML\HW\hw2.dat", unpack=True)

A = np.array([xm,np.ones(len(xm))]).T
xT = A.transpose()
out1 = np.dot(xT,ym)
out = np.dot(xT,A)
out2 = inv(out)
m,c = np.dot(out2,out1)
P1 = [m,c]


print("The Parameters of the first degree polynomial are : ")
print(P1)

XP = np.dot(A,P1)
error = (ym - XP)**2
optimalerror = error.min()
print("Optimal Error of degree 1 Polynomial: "+ str(optimalerror))

plt.plot(xm,ym,'.',label='Degree 1')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(xm, m*xm + c, '-', label='Regression line')
plt.legend(loc='best')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from numpy.linalg import inv

xm,ym = np.loadtxt("C:\Semester 2\ML\HW\hw2.dat", unpack=True)

A = np.array([xm**7,xm**6,xm**5,xm**4,xm**3,xm**2,xm, np.ones(len(xm))]).T
xT = A.transpose()
out1 = np.dot(xT,ym)
out = np.dot(xT,A)
out2 = inv(out)
a,b,c,d,e,f,g,h = np.dot(out2,out1)
P7=[a,b,c,d,e,f,g,h]

print("The Parameters of the Seventh degree polynomial are : ")
print(P7)



XP = np.dot(A,P7)
error = (ym - XP)**2
optimalerror = error.min()
print("Optimal Error of degree 7 Polynomial: "+ str(optimalerror))


plt.plot(xm,ym,'.',label='Degree 7')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(xm, a*(xm)**7 + b*(xm)**6 + c*(xm)**5 + d*(xm)**4 + e*(xm)**3
         + f*(xm)**2 + g*(xm) + h, '-', label='Regression line')
plt.legend(loc='best')
plt.show()
```
